```python
url = "../datasets/africa_food_prices.csv"

data = pd.read_csv(url)

data
```

```python
df = data.copy()
df
```

```python
cols_to_drop = ['Unnamed: 0']
df = df.drop(columns=cols_to_drop, axis=0)
```

```python
# 2.
# get the number of columns and rows

df.shape
```

```python
# 3.
# get the structure of the dataset - columns, non-null entries and data type.

df.info()
```

```python
# 4.
# Get the names of the columns

df.columns
```

```python
# 5.
# Find the  null entries

df.isnull().sum()
```

```python
# 6.
# dropped the last column having empty entries

cols_to_drop = ['mp_commoditysource']
df = df.drop(columns=cols_to_drop, axis=-1)
```

```python
# 7.
# dropped empty entries in the state column

df = df.dropna(subset=['state'])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# 8.
# Get duplicates

df.duplicated().sum()
```

```python
cols_to_drop = ['country_id', 'state_id','market_id','produce_id', 'currency_id', 'pt_id', 'um_unit_id']
df = df.drop(columns=cols_to_drop, axis=1)
```

```python
df.dtypes
```

```python
df['month'] = pd.to_datetime(df['month'], format='%m').dt.month_name().str.slice(stop=3)
df
```

```python
df['year'] = pd.to_datetime(df['year'], format='%Y').dt.year
df
```

```python
Nigeria = df.loc[df['country'] == 'Nigeria']
Nigeria_sort = Nigeria.sort_values('price')
Nigeria_sort
```

```python
Nigeria_sort['produce'].unique()
```
